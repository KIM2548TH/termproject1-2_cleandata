```python
import pandas as pd
import matplotlib.pyplot as plt

# อ่านข้อมูลจากไฟล์ CSV
df = pd.read_csv('province/data_eng_day.csv')

# ลบคอลัมน์ที่ไม่มีชื่อ (Unnamed)
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]

# แปลงคอลัมน์ timestamp ให้เป็นวันที่โดยเอาเวลาออก
df['timestamp'] = pd.to_datetime(df['timestamp']).dt.date

# ฟังก์ชันเพื่อกำหนดฤดูกาลตามวันที่
def get_season(date):
    if date.month in [5, 6, 7, 8, 9, 10]:
        return 'Rainy'
    else:
        return 'Summer'

# เพิ่มคอลัมน์ season ลงใน DataFrame
df['season'] = df['timestamp'].apply(get_season)

# ตรวจสอบคอลัมน์ใน DataFrame
print(df.columns)

# ตรวจสอบข้อมูลในคอลัมน์ season
print(df['season'].unique())

# ตรวจสอบข้อมูลใน DataFrame
print(df.head())

# สร้างคอลัมน์ใหม่สำหรับแต่ละฤดูกาล
df['Rainy'] = df['season'].apply(lambda x: 1 if x == 'Rainy' else 0)
df['Summer'] = df['season'].apply(lambda x: 1 if x == 'Summer' else 0)

# เลือกเฉพาะคอลัมน์ที่เป็นตัวเลข
numeric_columns = df.select_dtypes(include=['number']).columns

# แสดงกราฟตามฤดูกาล
seasonal_data = df.groupby('season')[numeric_columns].mean()

# ตรวจสอบข้อมูลใน seasonal_data
print(seasonal_data)

# พลอตกราฟแสดงข้อมูลตามฤดูกาล
y_columns = ['humidity', 'pm_10', 'pm_2_5', 'temperature']
seasonal_data[y_columns].plot(kind='bar', figsize=(10, 6))
plt.xlabel('Season')
plt.ylabel('Average Values')
plt.title('Average Values by Season')
plt.legend()
plt.grid(True)
plt.tight_layout()

# แสดงกราฟ
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# อ่านไฟล์ CSV
df = pd.read_csv('daily_csv/export-jsps014-1d.csv')

# ลบคอลัมน์ที่ไม่ต้องการ (Unnamed)
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]

# ฟังก์ชันเพื่อกำหนดฤดูกาลตามวันที่
def get_season(date):
    month = date.month
    if month >= 5 and month <= 10:
        return 'rainy'
    else:
        return 'summer'

# แปลงคอลัมน์ timestamp เป็น datetime
df['timestamp'] = pd.to_datetime(df['timestamp'])

# สร้างคอลัมน์ฤดูกาล
df['season'] = df['timestamp'].apply(get_season)

# แยกคอลัมน์ฤดูกาลเป็น 2 คอลัมน์ (rainy และ summer)
df['rainy'] = df['season'].apply(lambda x: 1 if x == 'rainy' else 0)
df['summer'] = df['season'].apply(lambda x: 1 if x == 'summer' else 0)

# แสดง DataFrame ใน Jupyter Notebook
print(df)

# พล้อตกราฟเส้นที่มีแกน x เป็น timestamp สำหรับทุกคอลัมน์
plt.figure(figsize=(14, 7))
for column in df.columns:
    if column != 'timestamp':
        plt.plot(df['timestamp'], df[column], label=column)

plt.xlabel('Timestamp')
plt.ylabel('Values')
plt.title('Time Series Data')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

# อ่านไฟล์ CSV
df = pd.read_csv('daily_csv/export-jsps014-1d.csv')

# ลบคอลัมน์ที่ไม่ต้องการ (Unnamed)
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]

# แปลงคอลัมน์ timestamp เป็น datetime
df['timestamp'] = pd.to_datetime(df['timestamp'])

# สร้างคอลัมน์ฤดูกาล
def get_season(date):
    month = date.month
    if month >= 5 and month <= 10:
        return 'rainy'
    else:
        return 'summer'

df['season'] = df['timestamp'].apply(get_season)

# แยกคอลัมน์ฤดูกาลเป็น 2 คอลัมน์ (rainy และ summer)
df['rainy'] = df['season'].apply(lambda x: 1 if x == 'rainy' else 0)
df['summer'] = df['season'].apply(lambda x: 1 if x == 'summer' else 0)

# Deseasonalization
df.set_index('timestamp', inplace=True)
result = seasonal_decompose(df['pm_2_5'], model='additive', period=365)
deseasonalized = df['pm_2_5'] - result.seasonal

# Detrending
deseasonalized_diff = deseasonalized.diff().dropna()

# การใช้ฟีเจอร์เสริม (Exogenous Features)
exog_features = df[['rainy', 'summer']].loc[deseasonalized_diff.index]

# การเลือกโมเดลที่เหมาะสม
model = SARIMAX(deseasonalized_diff, exog=exog_features, order=(1, 1, 1), seasonal_order=(1, 1, 1, 12))
results = model.fit()

# ตรวจสอบว่าดัชนีมีค่า '2022-01-01' หรือไม่
if pd.to_datetime('2022-01-01') in deseasonalized_diff.index:
    start_date = pd.to_datetime('2022-01-01')
else:
    start_date = deseasonalized_diff.index[0]

# ประเมินผลและการทำนาย
pred = results.get_prediction(start=start_date, dynamic=False)
pred_ci = pred.conf_int()

# การแสดงผลข้อมูล
ax = deseasonalized_diff['2021':].plot(label='Observed')
pred.predicted_mean.plot(ax=ax, label='Forecast', alpha=.7, figsize=(14, 7))
ax.fill_between(pred_ci.index, pred_ci.iloc[:, 0], pred_ci.iloc[:, 1], color='k', alpha=.2)
ax.set_xlabel('Date')
ax.set_ylabel('pm_2_5')
plt.legend()
plt.show()

# ประเมินผล
y_forecasted = pred.predicted_mean
y_truth = deseasonalized_diff[start_date:]
mse = mean_squared_error(y_truth, y_forecasted)
print(f'The Mean Squared Error of our forecasts is {mse}')
```
