```python
import pandas as pd
import matplotlib.pyplot as plt

# อ่านข้อมูลจากไฟล์ CSV
df = pd.read_csv('province/data_eng_day.csv')

# ลบคอลัมน์ที่ไม่มีชื่อ (Unnamed)
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]

# แปลงคอลัมน์ timestamp ให้เป็นวันที่โดยเอาเวลาออก
df['timestamp'] = pd.to_datetime(df['timestamp']).dt.date

# ฟังก์ชันเพื่อกำหนดฤดูกาลตามวันที่
def get_season(date):
    if date.month in [12, 1, 2]:
        return 'Winter'
    elif date.month in [3, 4, 5]:
        return 'Spring'
    elif date.month in [6, 7, 8]:
        return 'Summer'
    elif date.month in [9, 10, 11]:
        return 'Autumn'

# เพิ่มคอลัมน์ season ลงใน DataFrame
df['season'] = df['timestamp'].apply(get_season)

# สร้างคอลัมน์ใหม่สำหรับแต่ละฤดูกาล
df['Winter'] = df['season'].apply(lambda x: 1 if x == 'Winter' else 0)
df['Spring'] = df['season'].apply(lambda x: 1 if x == 'Spring' else 0)
df['Summer'] = df['season'].apply(lambda x: 1 if x == 'Summer' else 0)
df['Autumn'] = df['season'].apply(lambda x: 1 if x == 'Autumn' else 0)

# ตั้งค่าแกน x และ y
x = df['timestamp']
y_columns = ['humidity', 'pm_10', 'pm_2_5', 'temperature']

# พลอตกราฟเส้นที่มีจุด
plt.figure(figsize=(10, 6))
for column in y_columns:
    plt.plot(x, df[column], marker='o', label=column)

# ตั้งค่าชื่อแกนและชื่อกราฟ
plt.xlabel('Date')
plt.ylabel('Values')
plt.title('Time Series Data')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

# แสดงกราฟ
plt.show()

# แสดง DataFrame ที่มีคอลัมน์ season และฤดูกาล
print(df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# อ่านข้อมูลจากไฟล์ CSV
df = pd.read_csv('province/data_eng_day.csv')

# ลบคอลัมน์ที่ไม่มีชื่อ (Unnamed)
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]

# แปลงคอลัมน์ timestamp ให้เป็นวันที่โดยเอาเวลาออก
df['timestamp'] = pd.to_datetime(df['timestamp']).dt.date

# ฟังก์ชันเพื่อกำหนดฤดูกาลตามวันที่
def get_season(date):
    if date.month in [5, 6, 7, 8, 9, 10]:
        return 'Rainy'
    else:
        return 'Summer'

# เพิ่มคอลัมน์ season ลงใน DataFrame
df['season'] = df['timestamp'].apply(get_season)

# ตรวจสอบคอลัมน์ใน DataFrame
print(df.columns)

# ตรวจสอบข้อมูลในคอลัมน์ season
print(df['season'].unique())

# ตรวจสอบข้อมูลใน DataFrame
print(df.head())

# สร้างคอลัมน์ใหม่สำหรับแต่ละฤดูกาล
df['Rainy'] = df['season'].apply(lambda x: 1 if x == 'Rainy' else 0)
df['Summer'] = df['season'].apply(lambda x: 1 if x == 'Summer' else 0)

# เลือกเฉพาะคอลัมน์ที่เป็นตัวเลข
numeric_columns = df.select_dtypes(include=['number']).columns

# แสดงกราฟตามฤดูกาล
seasonal_data = df.groupby('season')[numeric_columns].mean()

# ตรวจสอบข้อมูลใน seasonal_data
print(seasonal_data)

# พลอตกราฟแสดงข้อมูลตามฤดูกาล
y_columns = ['humidity', 'pm_10', 'pm_2_5', 'temperature']
seasonal_data[y_columns].plot(kind='bar', figsize=(10, 6))
plt.xlabel('Season')
plt.ylabel('Average Values')
plt.title('Average Values by Season')
plt.legend()
plt.grid(True)
plt.tight_layout()

# แสดงกราฟ
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# อ่านข้อมูลจากไฟล์ CSV
df = pd.read_csv('province/data_eng_hour.csv')

# แปลงคอลัมน์ timestamp ให้เป็น datetime
df['timestamp'] = pd.to_datetime(df['timestamp'])

# ตั้งค่าแกน x และ y
x = df['timestamp']
y_columns = ['humidity', 'pm_10', 'pm_2_5', 'temperature']

# พลอตกราฟเส้นที่มีจุด
plt.figure(figsize=(12, 8))
for column in y_columns:
    plt.plot_date(x, df[column], marker='o', linestyle='-', label=column, markersize=4)

# ตั้งค่าชื่อแกนและชื่อกราฟ
plt.xlabel('Date')
plt.ylabel('Values')
plt.title('Time Series Data')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

# แสดงกราฟ
plt.show()
```
