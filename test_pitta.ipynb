```python
import pandas as pd
import matplotlib.pyplot as plt

#  CSV
df = pd.read_csv('province/data_eng_day.csv')

#  timestamp 
df['timestamp'] = pd.to_datetime(df['timestamp']).dt.date

#  x  y
x = df['timestamp']
y_columns = ['humidity', 'pm_10', 'pm_2_5', 'temperature']

# 
plt.figure(figsize=(10, 6))
for column in y_columns:
    plt.plot(x, df[column], marker='o', label=column)

# 
plt.xlabel('Date')
plt.ylabel('Values')
plt.title('Time Series Data')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

# 
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

#  CSV
df = pd.read_csv('province/data_eng_day.csv')

#  (Unnamed)
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]

#  timestamp 
df['timestamp'] = pd.to_datetime(df['timestamp']).dt.date

# 
def get_season(date):
    if date.month in [12, 1, 2]:
        return 'Winter'
    elif date.month in [3, 4, 5]:
        return 'Spring'
    elif date.month in [6, 7, 8]:
        return 'Summer'
    elif date.month in [9, 10, 11]:
        return 'Autumn'

#  season  DataFrame
df['season'] = df['timestamp'].apply(get_season)

# 
df['Winter'] = df['season'].apply(lambda x: 1 if x == 'Winter' else 0)
df['Spring'] = df['season'].apply(lambda x: 1 if x == 'Spring' else 0)
df['Summer'] = df['season'].apply(lambda x: 1 if x == 'Summer' else 0)
df['Autumn'] = df['season'].apply(lambda x: 1 if x == 'Autumn' else 0)

#  x  y
x = df['timestamp']
y_columns = ['humidity', 'pm_10', 'pm_2_5', 'temperature']

# 
plt.figure(figsize=(10, 6))
for column in y_columns:
    plt.plot(x, df[column], marker='o', label=column)

# 
plt.xlabel('Date')
plt.ylabel('Values')
plt.title('Time Series Data')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

# 
plt.show()

#  DataFrame  season 
print(df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

#  CSV
df = pd.read_csv('province/data_eng_day.csv')

#  (Unnamed)
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]

#  timestamp 
df['timestamp'] = pd.to_datetime(df['timestamp']).dt.date

# 
def get_season(date):
    if date.month in [5, 6, 7, 8, 9, 10]:
        return 'Rainy'
    else:
        return 'Summer'

#  season  DataFrame
df['season'] = df['timestamp'].apply(get_season)

#  DataFrame
print(df.columns)

#  season
print(df['season'].unique())

#  DataFrame
print(df.head())

# 
df['Rainy'] = df['season'].apply(lambda x: 1 if x == 'Rainy' else 0)
df['Summer'] = df['season'].apply(lambda x: 1 if x == 'Summer' else 0)

# 
numeric_columns = df.select_dtypes(include=['number']).columns

# 
seasonal_data = df.groupby('season')[numeric_columns].mean()

#  seasonal_data
print(seasonal_data)

# 
y_columns = ['humidity', 'pm_10', 'pm_2_5', 'temperature']
seasonal_data[y_columns].plot(kind='bar', figsize=(10, 6))
plt.xlabel('Season')
plt.ylabel('Average Values')
plt.title('Average Values by Season')
plt.legend()
plt.grid(True)
plt.tight_layout()

# 
plt.show()
```
